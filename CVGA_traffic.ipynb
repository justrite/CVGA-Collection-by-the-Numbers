```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
row_names = ['Traffic','Circulation']
traffic_2008_df = pd.read_csv('traffic/2008.csv')
traffic_2008_df.index = row_names
traffic_2009_df = pd.read_csv('traffic/2009.csv')
traffic_2009_df.index = row_names
traffic_2010_df = pd.read_csv('traffic/2010.csv')
traffic_2010_df.index = row_names
traffic_2011_df = pd.read_csv('traffic/2011.csv')
traffic_2011_df.index = row_names
traffic_2012_df = pd.read_csv('traffic/2012.csv')
traffic_2012_df.index = row_names
traffic_2013_df = pd.read_csv('traffic/2013.csv')
traffic_2013_df.index = row_names
traffic_2014_df = pd.read_csv('traffic/2014.csv')
traffic_2014_df.index = row_names
traffic_2015_df = pd.read_csv('traffic/2015.csv')
traffic_2015_df.index = row_names
traffic_2016_df = pd.read_csv('traffic/2016.csv')
traffic_2016_df.index = row_names
traffic_2017_df = pd.read_csv('traffic/2017.csv')
traffic_2017_df.index = row_names
traffic_2018_df = pd.read_csv('traffic/2018.csv')
traffic_2018_df.index = row_names
traffic_2019_df = pd.read_csv('traffic/2019.csv')
traffic_2019_df.index = row_names
```

```python
traffic_data = {2008:traffic_2008_df, 2009:traffic_2009_df, 2010:traffic_2010_df, 
        2011:traffic_2011_df, 2012:traffic_2012_df, 2013:traffic_2013_df, 
        2014:traffic_2014_df, 2015:traffic_2015_df, 2016:traffic_2016_df, 
        2017:traffic_2017_df, 2018:traffic_2018_df, 2019:traffic_2019_df}
```

```python
xlist = [2015, 2016, 2017, 2018, 2019]
ylist = []
for year in traffic_data:
    if year > 2014:
        ylist.append((traffic_data[year]['Totals']['Circulation'])/(traffic_data[year]['Totals']['Traffic']))
plt.title('CVGA Circulations by Year', size = 16, color='#00274c')
plt.xlabel('Year')
plt.ylabel('# of Loans per Visitor')
plt.xticks(xlist)
plt.plot(xlist, ylist, marker = 'o', color = '#00274c', markerfacecolor = '#ffcb05', 
         linewidth = 2.5, markersize = 10, markeredgewidth = 0)
plt.show()
```

```python
xlist = ['January', 'February', 'March', 'April']
cols = ['Jan', 'Feb', 'Mar', 'Apr']
ylist = []
for month in cols:
    ylist.append((traffic_data[2019][month]['Traffic']))
plt.title('CVGA Visitors in Winter 2019', size = 16, color='#00274c')
plt.xlabel('Month')
plt.ylabel('# of Visitors')
plt.plot(xlist, ylist, marker = 'o', color = '#00274c', markerfacecolor = '#ffcb05', 
         linewidth = 2.5, markersize = 10, markeredgewidth = 0)
plt.show()
```

```python
xlist = ['January', 'February', 'March', 'April']
cols = ['Jan', 'Feb', 'Mar', 'Apr']
ylist = []
for month in cols:
    ylist.append((traffic_data[2019][month]['Circulation']))
plt.title('CVGA Circulations in Winter 2019', size = 16, color='#00274c')
plt.xlabel('Month')
plt.ylabel('# of Loans')
plt.plot(xlist, ylist, marker = 'o', color = '#00274c', markerfacecolor = '#ffcb05', 
         linewidth = 2.5, markersize = 10, markeredgewidth = 0)
plt.show()
```
