```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
quantity_df = pd.read_csv("collection/collection_quantity.csv")
systems_df = pd.read_csv("collection/games_by_system.csv")
board_df = pd.read_csv("collection/board_games.csv")
```

```python
gen_2 = np.sum(systems_df[systems_df["Generation"] == 2]["Physical Games"]) + np.sum(systems_df[systems_df["Generation"] == 2]["Virtual Games"])
gen_3 = np.sum(systems_df[systems_df["Generation"] == 3]["Physical Games"]) + np.sum(systems_df[systems_df["Generation"] == 3]["Virtual Games"])
gen_4 = np.sum(systems_df[systems_df["Generation"] == 4]["Physical Games"]) + np.sum(systems_df[systems_df["Generation"] == 4]["Virtual Games"])
gen_5 = np.sum(systems_df[systems_df["Generation"] == 5]["Physical Games"]) + np.sum(systems_df[systems_df["Generation"] == 5]["Virtual Games"])
gen_6 = np.sum(systems_df[systems_df["Generation"] == 6]["Physical Games"]) + np.sum(systems_df[systems_df["Generation"] == 6]["Virtual Games"])
gen_7 = np.sum(systems_df[systems_df["Generation"] == 7]["Physical Games"]) + np.sum(systems_df[systems_df["Generation"] == 7]["Virtual Games"])
gen_8 = np.sum(systems_df[systems_df["Generation"] == 8]["Physical Games"]) + np.sum(systems_df[systems_df["Generation"] == 8]["Virtual Games"])
gens = [gen_8, gen_7, gen_6, gen_5, gen_4, gen_3, gen_2]
gen_labels = ["8th (2012-Present)", "7th (2005-2017)", "6th (1998-2013)", 
              "5th (1993-2005)", "4th (1987-2004)", "3rd (1983-2003)", "2nd (1976-1992)"]
colors = ['#00274c', '#ffcb05','#00274c', '#ffcb05','#00274c', '#ffcb05','#00274c']
```

```python
plt.barh(gen_labels, gens, color = colors)
plt.title('Number of CVGA Games by Console Generation', size = 16, color='#00274c')
plt.ylabel('Console Generation', size = 12)
plt.xlabel('Number of Games', size = 12)
plt.text(0, -2.5, '*does not include PC/board games', fontsize = 8)
plt.text(650, -2.5, 'https://en.wikipedia.org/wiki/Video_game_console', fontsize = 8)
plt.show()
```

```python
pc_physical = systems_df[systems_df["Console"] == "PC"]["Physical Games"].item()
pc_virtual = systems_df[systems_df["Console"] == "PC"]["Virtual Games"].item()
labels = 'Physical Copies', 'Virtual Copies'
sizes = [pc_physical, pc_virtual]
colors = '#ffcb05', '#00274c'
plt.title("Format of CVGA PC Games", size = 16, color='#00274c')
plt.pie(x = sizes, explode = (0.05, 0), startangle = 90, labels=labels, colors = colors, 
        autopct='%1.1f%%', textprops = dict(fontsize = 12))
plt.show()
```

```python
quantity_2008 = quantity_df[quantity_df["Year"]==2008]["Physical"].item() + quantity_df[quantity_df["Year"]==2008]["Virtual"].item()
quantity_2009 = quantity_df[quantity_df["Year"]==2009]["Physical"].item() + quantity_df[quantity_df["Year"]==2009]["Virtual"].item()
quantity_2010 = quantity_df[quantity_df["Year"]==2010]["Physical"].item() + quantity_df[quantity_df["Year"]==2010]["Virtual"].item()
quantity_2011 = quantity_df[quantity_df["Year"]==2011]["Physical"].item() + quantity_df[quantity_df["Year"]==2011]["Virtual"].item()
quantity_2012 = quantity_df[quantity_df["Year"]==2012]["Physical"].item() + quantity_df[quantity_df["Year"]==2012]["Virtual"].item()
quantity_2013 = quantity_df[quantity_df["Year"]==2013]["Physical"].item() + quantity_df[quantity_df["Year"]==2013]["Virtual"].item()
quantity_2014 = quantity_df[quantity_df["Year"]==2014]["Physical"].item() + quantity_df[quantity_df["Year"]==2014]["Virtual"].item()
quantity_2015 = quantity_df[quantity_df["Year"]==2015]["Physical"].item() + quantity_df[quantity_df["Year"]==2015]["Virtual"].item()
quantity_2016 = quantity_df[quantity_df["Year"]==2016]["Physical"].item() + quantity_df[quantity_df["Year"]==2016]["Virtual"].item()
quantity_2017 = quantity_df[quantity_df["Year"]==2017]["Physical"].item() + quantity_df[quantity_df["Year"]==2017]["Virtual"].item()
quantity_2018 = quantity_df[quantity_df["Year"]==2018]["Physical"].item() + quantity_df[quantity_df["Year"]==2018]["Virtual"].item()
quantity_years = {2008:quantity_2008, 2009:quantity_2009, 2010:quantity_2010, 2011:quantity_2011, 2012:quantity_2012, 2013:quantity_2013,
                  2014:quantity_2014, 2015:quantity_2015, 2016:quantity_2016, 2017:quantity_2017, 2018:quantity_2018}    
```

```python
xlist = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]
ylist = []
for year in quantity_years:
    ylist.append(quantity_years[year])
plt.title('Size of CVGA Collection by Year', size = 16, color='#00274c')
plt.xlabel('Year')
plt.ylabel('Number of Games in the CVGA Collection')
plt.xticks(xlist)
plt.plot(xlist, ylist, marker = 'o', color = '#00274c', markerfacecolor = '#ffcb05', linewidth = 2.5, markersize = 5, markeredgewidth = 0.5)
plt.show()
```
